```stata

qui reg coneduc i.race i.age educ i.sex, ro

quietly margins i.race, atmeans 

marginsplot, recast(scatter) scheme(dubois) ytitle(Confidence in Education) ///
title("Figure 3: Confidence in Education by Race" " ") ///
```
