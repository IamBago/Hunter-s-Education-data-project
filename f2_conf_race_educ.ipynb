```stata
use race educ coneduc age sex racdif3 using GSS2022-Copy1.dta, clear
```

```stata
%set graph_width = 11
```

```stata
%set graph_height = 8
```

```stata
label values educ drop
```

```stata
quietly reg coneduc c.educ##i.race age i.sex

quietly margins i.race, at(educ=(0(1)20)) atmeans

marginsplot,  xtitle(" " "Years in Education") xlabel(0(5)20) ///
    title("Figure 2: Confidence in Education by Race and Education" " ") ///
    ytitle("Confidence in Education") legend(rows(3) position(3)) scheme(dubois)
```
