```stata
use race educ coneduc age sex racdif3 using GSS2022-Copy1.dta, clear
```

```stata
%set graph_width = 11
```

```stata
%set graph_height = 8
```

```stata
quietly reg coneduc c.educ##i.race c.age##i.sex##c.educ
quietly margins i.race, over(coneduc) atmeans
marginsplot, recast(scatter) scheme(538w)
```

```stata
quietly reg coneduc c.educ##i.race##c.age##i.sex##c.educ

quietly margins race, at(educ=(0(1)20)) atmeans 
marginsplot, noci legend(on) xtitle("Years in Education") ///
ytitle("Confidence in Education") legend(rows(1) position(1) ring(0)) scheme (538w)
```
